# Merging all tables to create unique IDs

```python
pd.set_option('display.max_rows', None)

df_merging = pd.concat([data_andreu_butacas, data_andreu_sillas_sillones, data_doos_butacas, data_doos_sillas_sillones, data_inclass_sillas, data_inclass_sillones_butacas,data_ondarreta, data_viccarbe_butacas, data_viccarbe_sillas_sillones ], ignore_index=True)
df_merging.head()
```

# Creation of IDs
### Product

```python
df_merging['Product_id'] = range(1, len(df_merging)+1)
df_merging.head()
```

### Category

```python
category_to_id = {
    'Sillón': 1,
    'Silla': 2,
    'Butaca': 3,
    'Reposapiés': 4,
    'Módulo': 5,
    'Banco': 6,
    'Pouf': 7,
    'Sofá': 8,
    'Taburete': 9,
    'Reposapies': 10
}
df_merging['Category_id'] = df_merging['Category'].map(category_to_id)
df_merging.head()
```

### Company

```python
company_to_id = {
    'Andreu World': 1,
    'Doos': 2,
    'Inclass': 3,
    'Viccarbe': 4,
    'Ondarreta': 5
}

df_merging['Company_id'] = df_merging['Company'].map(company_to_id)
df_merging.head()


```

# Creation of categorized tables and excels for use on Power BI

```python
df_category_id = df_merging[['Category', 'Category_id']]
df_category_id.drop_duplicates(keep='first', ignore_index=True, inplace=True)

df_product_id = df_merging[['Product', 'Product_id']]

df_company_id = df_merging[['Company', 'Company_id']]
df_company_id.drop_duplicates(keep='first', ignore_index=True, inplace=True)

df_combined_id = df_merging[['Category_id', 'Company_id', 'Product_id']]
df_combined_id.drop_duplicates(keep='first', ignore_index=True, inplace=True)


with pd.ExcelWriter('Category id.xlsx') as writer:
    df_category_id.to_excel(writer, sheet_name='Category')

with pd.ExcelWriter('Product id.xlsx') as writer:
    df_product_id.to_excel(writer, sheet_name='Product')

with pd.ExcelWriter('Company id.xlsx') as writer:
    df_company_id.to_excel(writer, sheet_name='Company')

with pd.ExcelWriter('Combined id.xlsx') as writer:
    df_combined_id.to_excel(writer, sheet_name='Combined')

with pd.ExcelWriter('Merged.xlsx') as writer:
    df_merging.to_excel(writer, sheet_name='Merged')
```
