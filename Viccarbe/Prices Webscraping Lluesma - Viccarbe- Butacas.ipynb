```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# Web Scraping

```python
url_lluesma = "https://www.muebleslluesma.com/2189-butacas-viccarbe"
page_lluesma = urlopen(url_lluesma)

soup_lluesma = bs(page_lluesma, 'html.parser')
print(soup_lluesma.prettify().splitlines()[0:30])

names_lluesma = soup_lluesma.find_all('h3', class_='h3 product-title')

for name in names_lluesma[0:20]:
    print(name.prettify())
    
description_lluesma = soup_lluesma.find_all('div', class_='product-description-short')

for description in description_lluesma[0:20]:
    print(description.prettify())

regular_prices_lluesma = soup_lluesma.find_all('span', class_='regular-price')

for regular_prices in regular_prices_lluesma[0:20]:
    print(regular_prices.prettify())
```

```python
titles = []
descriptions = []
prices = []

for i in range(len(names_lluesma)):
    title = names_lluesma[i].text if names_lluesma[i] else None
    description = description_lluesma[i].text.strip() if description_lluesma[i] else None
    price = regular_prices_lluesma[i].text.replace('\n', '') if regular_prices_lluesma[i] else None
    
    titles.append(title)
    descriptions.append(description)
    prices.append(price)

df_lluesma = pd.DataFrame({'Title': titles, 'Description': descriptions, 'Price': prices})
```

# Data Frame Manipulation
### Merging all pages

```python
pd.set_option('display.max_rows', None)

df_combined_viccarbe_butacas = df_lluesma
df_combined_viccarbe_butacas.head(100)
```

### Extraction of infromation from 'Title' column

```python
df_combined_viccarbe_butacas.dtypes
```

```python
df_combined_viccarbe_butacas[['Category','Name']] = df_combined_viccarbe_butacas['Title'].str.split(' ', n=1, expand=True)
df_combined_viccarbe_butacas
```

### Deletion of repeated and unnecessary info

```python
df_combined_viccarbe_butacas['Name'] = df_combined_viccarbe_butacas['Name'].str.replace('Viccarbe','')
df_combined_viccarbe_butacas
```

### Deletion of unwanted columns

```python
df_combined_viccarbe_butacas.drop(['Title'], axis=1, inplace=True)
```

```python
with pd.ExcelWriter('Viccarbe-Butacas.xlsx') as writer:
    df_combined_viccarbe_butacas.to_excel(writer, sheet_name='Viccarbe-Butacas')
```
