```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# Web Scraping

### Page 1

```python
url_lluesma = "https://www.muebleslluesma.com/2215-butacas-andreu-world"
page_lluesma = urlopen(url_lluesma)

soup_lluesma = bs(page_lluesma, 'html.parser')
print(soup_lluesma.prettify().splitlines()[0:30])

names_lluesma = soup_lluesma.find_all('h3', class_='h3 product-title')

for name in names_lluesma[0:20]:
    print(name.prettify())
    
description_lluesma = soup_lluesma.find_all('div', class_='product-description-short')

for description in description_lluesma[0:20]:
    print(description.prettify())

prices_lluesma = soup_lluesma.find_all('span', class_='price')

for prices in prices_lluesma[0:20]:
    print(prices.prettify())
    
```

```python
titles = []
descriptions = []
prices = []

for i in range(len(names_lluesma)):
    title = names_lluesma[i].text if names_lluesma[i] else None
    description = description_lluesma[i].text.strip() if description_lluesma[i] else None
    price = prices_lluesma[i].text.replace('\n', '') if prices_lluesma[i] else None
    
    titles.append(title)
    descriptions.append(description)
    prices.append(price)

df_lluesma = pd.DataFrame({'Title': titles, 'Description': descriptions, 'Price': prices})

df_lluesma.head(30)
```

# Data Frame Manipulation
### Combination of all pages

```python
pd.set_option('display.max_rows', None)

df_combined_andreu_butacas = pd.concat([df_lluesma, df_lluesma_page2, df_lluesma_page3, df_lluesma_page4], ignore_index=True)
df_combined_andreu_butacas.head(100)
```

### Extraction of infromation from 'Title' column

```python
df_combined_andreu_butacas.dtypes
```

```python
df_combined_andreu_butacas['Title'] = df_combined_andreu_butacas['Title'].astype(str)
```

```python
df_combined_andreu_butacas.dtypes
```

```python
df_combined_andreu_butacas[['Category','Name']] = df_combined_andreu_butacas['Title'].str.split(' ', n=1, expand=True)
df_combined_andreu_butacas
```

### Deletion of repeated and unnecessary info

```python
df_combined_andreu_butacas['Name'] = df_combined_andreu_butacas['Name'].str.replace('Andreu World','')
df_combined_andreu_butacas
```

### Deletion of unwanted columns

```python
df_combined_andreu_butacas.drop(['Title'], axis=1, inplace=True)
```

```python
df_combined_andreu_butacas.head()
```

# Exporting IPYNB 'Jupyter' format to XLSX 'Excel'

```python
with pd.ExcelWriter('Andreu-Butacas.xlsx') as writer:
    df_combined_andreu_butacas.to_excel(writer, sheet_name='Andreu-Butacas')
```
