```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

# Web Scraping
### Page 1 / 2 / 3

```python
#Page1#
url_lluesma = "https://www.muebleslluesma.com/2518-sillas-inclass"
page_lluesma = urlopen(url_lluesma)

soup_lluesma = bs(page_lluesma, 'html.parser')
print(soup_lluesma.prettify().splitlines()[0:30])

names_lluesma = soup_lluesma.find_all('h3', class_='h3 product-title')

for name in names_lluesma[0:20]:
    print(name.prettify())
    
description_lluesma = soup_lluesma.find_all('div', class_='product-description-short')

for description in description_lluesma[0:20]:
    print(description.prettify())

regular_prices_lluesma = soup_lluesma.find_all('span', class_='regular-price')

for regular_prices in regular_prices_lluesma[0:20]:
    print(regular_prices.prettify())


#Page2#
url_lluesma_page2 = "https://www.muebleslluesma.com/2518-sillas-inclass?page=2"
page_lluesma_page2 = urlopen(url_lluesma_page2)

soup_lluesma_page2 = bs(page_lluesma_page2, 'html.parser')
print(soup_lluesma_page2.prettify().splitlines()[0:30])

names_lluesma_page2 = soup_lluesma_page2.find_all('h3', class_='h3 product-title')

for name_page2 in names_lluesma_page2[0:20]:
    print(name_page2.prettify())
    
description_lluesma_page2 = soup_lluesma_page2.find_all('div', class_='product-description-short')

for description_page2 in description_lluesma_page2[0:20]:
    print(description_page2.prettify())

regular_prices_lluesma_page2 = soup_lluesma_page2.find_all('span', class_='regular-price')

for regular_prices_page2 in regular_prices_lluesma_page2[0:20]:
    print(regular_prices_page2.prettify())


#Page3#
url_lluesma_page3 = "https://www.muebleslluesma.com/2518-sillas-inclass?page=3"
page_lluesma_page3 = urlopen(url_lluesma_page3)

soup_lluesma_page3 = bs(page_lluesma_page3, 'html.parser')
print(soup_lluesma_page3.prettify().splitlines()[0:30])



names_lluesma_page3 = soup_lluesma_page3.find_all('h3', class_='h3 product-title')

for name_page3 in names_lluesma_page3[0:20]:
    print(name_page3.prettify())
    
description_lluesma_page3 = soup_lluesma_page3.find_all('div', class_='product-description-short')

for description_page3 in description_lluesma_page3[0:20]:
    print(description_page3.prettify())

regular_prices_lluesma_page3 = soup_lluesma_page3.find_all('span', class_='regular-price')

for regular_prices_page3 in regular_prices_lluesma_page3[0:20]:
    print(regular_prices_page3.prettify())
```

```python
#Page1#
titles = []
descriptions = []
prices = []

for i in range(len(names_lluesma)):
    title = names_lluesma[i].text if names_lluesma[i] else None
    description = description_lluesma[i].text.strip() if description_lluesma[i] else None
    price = regular_prices_lluesma[i].text.replace('\n', '') if regular_prices_lluesma[i] else None
    
    titles.append(title)
    descriptions.append(description)
    prices.append(price)

df_lluesma = pd.DataFrame({'Title': titles, 'Description': descriptions, 'Price': prices})

#Page2#
titles = []
descriptions = []
prices = []

for i in range(len(names_lluesma_page2)):
    title = names_lluesma_page2[i].text if names_lluesma_page2[i] else None
    description = description_lluesma_page2[i].text.strip() if description_lluesma_page2[i] else None
    price = regular_prices_lluesma_page2[i].text if regular_prices_lluesma_page2[i] else None
    
    titles.append(title)
    descriptions.append(description)
    prices.append(price)

df_lluesma_page2 = pd.DataFrame({'Title': titles, 'Description': descriptions, 'Price': prices})

#Page3#
titles = []
descriptions = []
prices = []

for i in range(len(names_lluesma_page3)):
    title = names_lluesma_page3[i].text if names_lluesma_page3[i] else None
    description = description_lluesma_page3[i].text.strip() if description_lluesma_page3[i] else None
    price = regular_prices_lluesma_page3[i].text if regular_prices_lluesma_page3[i] else None
    
    titles.append(title)
    descriptions.append(description)
    prices.append(price)

df_lluesma_page3 = pd.DataFrame({'Title': titles, 'Description': descriptions, 'Price': prices})
```

# Data Frame Manipulation
### Merging all pages

```python
pd.set_option('display.max_rows', None)

df_combined_inclass_sillas = pd.concat([df_lluesma, df_lluesma_page2, df_lluesma_page3], ignore_index=True)
df_combined_inclass_sillas.head(200)
```

### Extraction of infromation from 'Title' column

```python
df_combined_inclass_sillas.dtypes
```

```python
df_combined_inclass_sillas[['Category','Product']] = df_combined_inclass_sillas['Title'].str.split(' ', n=1, expand=True)
df_combined_inclass_sillas
```

```python
df_combined_inclass_sillas['Company'] = 'Inclass'
df_combined_inclass_sillas
```

### Deletion of repeated and unnecessary info

```python
df_combined_inclass_sillas['Product'] = df_combined_inclass_sillas['Product'].str.replace('Inclass','')
df_combined_inclass_sillas
```

### Deletion of unwanted columns

```python
df_combined_inclass_sillas.drop(['Title'], axis=1, inplace=True)
```
